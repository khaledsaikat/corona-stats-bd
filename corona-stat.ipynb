# Corona Stats
Corona stats in Bangladesh

```python
# Import libs
import matplotlib.pyplot as plt
import json
```

```python
def graph(file_name, y_axis):
    '''Plot the graph from a json file'''
    with open('data/' + file_name) as json_file:
        data = json.load(json_file)
        x = [itm['date'] for itm in data]
        y = [itm[y_axis] for itm in data]
        plt.plot(x, y)
        plt.xlabel('date')
        plt.ylabel(y_axis)
        plt.xticks(x, x, rotation='vertical')
        plt.show()
```

## Cases 24hr

```python
graph('iedcr_test_data_cases_24hr_3004_0406.json', 'cases_24hr')
```
